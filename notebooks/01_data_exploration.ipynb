# Transactions data exploration

The transactions dataset contains individual financial operations. Each transaction belongs to one account and is associated with one merchant. This table connects the accounts and merchants datasets

```python
transactions_path = (project_root / "data" / "raw" / "transactions.csv")
transactions_path
```

```python
transactions_path.exists()
```

```python
transactions = pd.read_csv(transactions_path)
transactions.head()
```

```python
transactions.tail()
```

```python
transactions.shape
```

```python
transactions.columns
```

```python
transactions.dtypes
```

```python
# Converting timestamp column to datetime format
transactions["timestamp"] = pd.to_datetime(transactions["timestamp"],
                                           format = "%Y-%m-%d %H:%M:%S")
```

```python
transactions["timestamp"].dtype
```

```python
transaction_period = {
    "first_transaction": transactions["timestamp"].min(),
    "last_transaction": transactions["timestamp"].max(),
}

transaction_period
```

```python
transactions["amount"].describe()
```

```python
transactions["status"].value_counts()
```

```python
transactions["transaction_type"].value_counts()
```

```python
transactions["currency"].value_counts()
```

```python
completed_transactions = transactions.loc[transactions["status"] == "Completed"]
completed_transactions.head()
```

```python
completed_transactions.shape
```

```python
declined_transactions = transactions.loc[transactions["status"] == "Declined"]
declined_transactions
```

```python
transactions.loc[transactions["amount"] > 100, ["transaction_id", "timestamp", "amount", "currency", "status"]]
```

```python
transactions.loc[transactions["amount"].between(80,100)]
```

```python
transactions.loc[
    (transactions["status"] == "Completed") & (transactions["amount"] > 100)
]
```

```python
february_start = pd.Timestamp("2025-02-01")
march_start = pd.Timestamp("2025-03-01")
```

```python
february_transactions = transactions.loc[(transactions["timestamp"] >= february_start)
                                         & (transactions["timestamp"] < march_start)
                                         ]
february_transactions
```

```python
transactions.sort_values("timestamp", ascending = False).head()
```

```python
transactions["transaction_id"].is_unique
```

```python
transactions["transaction_id"].nunique() == len(transactions)
```

```python
transactions.isna().sum()
```

```python
valid_account_references = transactions["account_id"].isin(accounts["account_id"])
valid_account_references.all()
```

```python
invalid_account_transactions = transactions.loc[~transactions["account_id"].isin(accounts["account_id"])]
invalid_account_transactions
```

```python
invalid_account_transactions.empty
```

```python
valid_merchant_references = (
    transactions["merchant_id"]
    .isin(merchants["merchant_id"])
)

valid_merchant_references.all()
```

```python
(transactions["amount"] > 0).all()
```

```python
# Summary
transaction_summary = {
    "number_of_transactions": len(transactions),
    "completed_transactions": (
        transactions["status"] == "Completed"
    ).sum(),
    "declined_transactions": (
        transactions["status"] == "Declined"
    ).sum(),
    "purchase_transactions": (
        transactions["transaction_type"] == "Purchase"
    ).sum(),
    "refund_transactions": (
        transactions["transaction_type"] == "Refund"
    ).sum(),
    "average_amount": transactions["amount"].mean(),
    "maximum_amount": transactions["amount"].max(),
    "minimum_amount": transactions["amount"].min(),
}

transaction_summary
```

## Initial transactions findings

- The dataset contains 60 transactions between January and March 2025
- Most transactions are completed purchases
- Three transactions were declined
- One transaction represents a refund
- Most transactions use EUR while a small number use USD
- Transaction identifiers are unique
- All account references exist in the accounts dataset
- All merchant references exist in the merchants dataset
- The dataset does not contain missing values
- All transaction amounts are greater than zero
